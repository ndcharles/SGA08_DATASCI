```python
# defining the dependent and independent variables
# let's see how the curb-weight is dependent on the engine-size
y = autos['curb-weight']
x1 = autos['engine-size']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# Let's first use statsmodel to perform the operation
# adding constants, x
x = sm.add_constant(x1)
# and result variable
results = sm.OLS(y, x).fit()
results.summary()
```

```python
# plotting the regression line
plt.scatter(x1, y)
yhat = 14.0596*x1 + 784.8662
fig = plt.plot(x1, yhat, lw=2, c='orange', label = 'regression line')
plt.xlabel('Engine Size', fontsize = 20)
plt.ylabel('Curb Weight', fontsize = 20)
plt.show()
```

```python
# divide dataset into attr (independent) and labels (dependent)
X = autos['engine-size'].values.reshape(-1,1)
y = autos['curb-weight'].values.reshape(-1,1)
```

```python
# we split data to train and test (80/20, resp)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#training the algorithm
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 
```

```python
#To retrieve the intercept:
print(regressor.intercept_)

#For retrieving the slope:
print(regressor.coef_)
```

So, the result for intercept and slope is 872.71682753 and 13.46542659 respectively.

y = 872.7168 + 13.4654x

This means that for every one unit of change in engine size, there is about 13.46% change in curb weight.

```python
# Let's use our test data and see how accurately our algorithm predicts the percentage score
y_pred = regressor.predict(X_test)
```

```python
# compare the actual output values for X_test with the predicted values
autos = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
```
