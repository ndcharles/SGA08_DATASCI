# African Economy Analysis
Our aim is to understand the relationship between *systemic crises*, *banking crises* and *African economy*. 

Financial stability is the ability of the financial system to facilitate and enhance economic processes, manage risks, and absorb shocks.

A **(systemic) banking crisis** occurs when many banks in a country are in serious solvency or liquidity problems at the same time. It is a general realization that systemically important financial institutions are in distress. This distress can lead to depressed asset prices, increase in interest rates, and a slowdown or reversal in capital flows.

The results of a World Bank Working Paper (1997) suggests that crises tend to erupt when the macroeconomic environment is weak. It reported that low GDP growth, excessively high real interest rates, and high inflation significantly increase the likelihood of systemic problems.

From the Pearson correlation, which is furthered ilustrated using the above heatmap, we can see that the systemic crisis is highly correlated to banking crisis (pearson coeff = 0.85). By definition, the occurence of a systemic banking crisis means that the economy is in a financial distress. This distress leads to an increase in interest rates and a high tendency of inflation in the economy.

To understand further, the effects of banking crisis and systemic crisis on the economy, let's look closely at such economic index as domestic debt in default, sovereign external debt default, inflation crisis. The inflation crisis however, is somewhat correlated to the currency crises (pearson coeff = 0.39). The year and independence might be another intriguing thing to look at.

That leaves one with the question: Does gaining independence lead to an economic crisis?
