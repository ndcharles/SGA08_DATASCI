```python
import seaborn as sns
```

# mpg vs. hp 

```python
# Determining the relationship between mpg and hp based on transmission (am)
sns.relplot(x="mpg", y="hp", hue="am", data=mtcars);
```

linear negative relation: as horse power of the engine (hp) increases, the mileage (mpg) reduces. Cars with manual transmission seems however to be more efficient.

# mpg vs wt

```python
# Relationship between mpg and weight based on transmission (am)
sns.relplot(x="mpg", y="wt", hue="am", data=mtcars);
```

mpg vs. weight: negative relation, the functional form might be non-linear, as weight of the car increases, the mileage decreases. The weight variable seems to provide perfect separation between manual and automatic transmission cars, i.e. all cars that are heavier than 3.2 ton (circa) are automatic and vice-versa.

```python
# Relatonship between mpg and drat based on transmission (am)
sns.relplot(x="mpg", y="drat", hue="am", data=mtcars);
```
