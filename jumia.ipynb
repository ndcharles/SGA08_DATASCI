**Scraping Jumia to view their catalogue of educational devices and their prices.** 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
import csv
```

```python
csv_file = open('jumia.csv', 'w', newline='', encoding='UTF-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['device_name', 'discount_price', 'former_price'])
```

```python
source = requests.get('https://www.jumia.com.ng/educational-tablets/').text
soup = BeautifulSoup(source, 'lxml')
main = soup.find_all('article', class_="prd _fb col c-prd")
for column in main:
    device_name = column.h3.text
    print(device_name)
    try:
        discount_price = column.find('div', class_='prc').text
        currency = discount_price[:1]
        price = discount_price[1:]
        print(price)
        print(currency)
    except Exception as e:
        pass
    try:
        former_price = column.find('div', class_='old').text
        currency1 = former_price[:1]
        price1 = former_price[1:]
        print(currency1)
        print(price1)
    except Exception as e:
        pass
    print()
    
    csv_writer.writerow([device_name, discount_price, former_price])

csv_file.close()
```
