# Merging two NBA datasets (nba list and salaries)

```python
import pandas as pd
```

```python
players = pd.read_csv('players.csv')
salary = pd.read_csv('salaries_1985to2018.csv')
```

```python
players.info()
print()
print()
salary.info()
```

```python
players.head()
```

```python
players.columns
```

```python
players.drop(['career_AST', 'career_FG%', 'career_FG3%', 'career_FT%', 'career_G', 'career_PER', 'career_PTS', 'career_TRB', 'career_WS', 'career_eFG%', 'draft_round'], inplace=True, axis=1)
```

```python
players.columns
```

```python
players.head()
```

```python
salary.columns
```

```python
salary.head()
```

```python
players_merged = players.merge(salary, left_on='_id', right_on='player_id')
```

```python
players_merged.head()
```

```python
players_merged.corr()
```

# Correlating NBA dataset

```python
NBA = pd.read_csv('nba.csv')
```

```python
NBA.head()
```

```python
NBA.info()
```

```python
NBA.dropna(inplace=True)
```

```python
NBA.head()
```

```python
# to convert positions to integer values
NBA['Position'].unique()
```

```python
# create a dictionary of key pairs
Position = {'PG': 1, 'SF': 2, 'SG': 3, 'PF': 4, 'C': 5}
NBA.Position = [Position[item] for item in NBA.Position]
print(Position)
```

```python
NBA.head(15)
```

```python
NBA.corr(method='pearson')
```

```python
NBA.corr(method='spearman')
```
